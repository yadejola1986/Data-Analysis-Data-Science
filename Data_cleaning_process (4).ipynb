```python
import re

def removeChar(txt):
    new_txt = re.sub(r'[ /!#$%&*()-]', "_", txt)
    return new_txt

df.columns = [removeChar(cols) for cols in df]
df.columns
```

```python
df= df.drop(['PK', 'Date','Farm_ID', 'Age_of_hens__week_','Farm_name', 'Farm_Site', 'Farm_site_number', 'State',
       'Location', 'Geographic_location', 'Total_mortality','Year', 'Week_day', 'Hours_of_light',
       'Data_source', 'Shed_name', 'Shed_number', 'Shed_type', 'Batch_number', 'Production_stage', 'Breed_name','Range_access_out',
       'Range_access_in', 'Loose_Birds_caught', 'Ammonium_level',
       'Mortality____Nestbox', 'Mortality___Ground', 'Mortality_hens___Range',
       'Mortality_hens___Cull',
       'Cummulative_mortality____', 'Total_feed_consumption_silo_A__kg_',
       'Total_feed_consumption_Silo_B__kg_','Type_of_ventilation', 'Clean_eggs',
       'System_eggs', 'Floor_eggs', 'Waste_eggs', 'Total_eggs',
       'Number_of_hens','Comments'], axis = 1)
```

```python
df.columns
```

```python
df.rename(columns={'Flock_ID':'flock_id', 'Age_of_hens__days_':'age', 'Mortality____':'mortality',
                   'Feed_consumption_hen_day__g_':'feed_intake','Water_consumption_hen_day__ml_':'water_intake','Indoor_Minimum_temperature__C_':'indoor_min_temp',
                   'Indoor_Maximum_temperature__C_':'indoor_max_temp', 'Laying_rate____':'laying_rate', 'Total_daily_precipitation__mm_':'precipitation',
       'Solar_radiation__MJ___meter_squared_per_day_':'solar_radiation',
                   'Outdoor_Minimum_temperature__C__':'outdoor_min_temp','Outdoor_Maximum_temperature__C__':'outdoor_max_temp',
                   'Vapour_pressure_9am__kilopascal_':'vp_9am', 'Vapour_pressure_3pm__kilopascal_':'vp_3pm',
                  '_Windspeed__m_s_':'windspeed', 'Relative_humidity':'rh_9am', 'Relative_humidity.1':'rh_3pm'}, inplace=True)
```

```python
# Strip spaces from column names
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
#Reset the flock_Id to a sequential numbering starting from 'flock 1'
df['flock_id'] = 'flock ' + (pd.factorize(df['flock_id'])[0] + 1).astype(str).astype(object)
```

```python
# Plotting egg production trends

plt.figure(figsize=(10, 5))

sns.scatterplot(x='age', y='laying_rate', hue='flock_id', data=df, marker='o')

plt.title('Egg production curve')
plt.xlabel('Hen Age(days)')
plt.ylabel('Laying rate(%)')
plt.grid()
plt.show()
```

```python
#Select rows where layingRate_avg are below 98
df= df.loc[df['laying_rate'] <= 98]
```

```python
##To check columns with null values
columns_with_nan = df.columns[df.isnull().any()].tolist()
print(columns_with_nan)
```

```python
# Replace '0' with NaN for proper averaging
df.replace(0, np.nan, inplace=True)
```

```python
# Replace NaN values with the local mean based on 'hen_age'
for col in df.columns[2:]:
    # Convert the column to numeric, coercing errors to NaN
    df[col] = pd.to_numeric(df[col], errors='coerce')

    # Check if the column is numeric after coercion
    if pd.api.types.is_numeric_dtype(df[col]):
        df[col] = df.groupby('age')[col].transform(lambda x: x.fillna(x.mean()))
    else:
        print(f"Skipping column '{col}' as it is not numeric.")
```

```python
def check_NaN(dataframe):
    print("Total NaN:", dataframe.isnull().values.sum())
    print("NaN by column:\n",dataframe.isnull().sum())
    return
```

```python
check_NaN(df)
```

```python
#Drop rows with any remaining NaN values
df.dropna(inplace=True)
```

```python
def check_NaN(dataframe):
    print("Total NaN:", dataframe.isnull().values.sum())
    print("NaN by column:\n",dataframe.isnull().sum())
    return
```

```python
check_NaN(df)
```

```python
# Plotting egg production trends
plt.figure(figsize=(10, 5))

sns.scatterplot(x='age', y='laying_rate', hue='flock_id', data=df, marker='o')

plt.title('Egg production curve')
plt.xlabel('Hen Age(days)')
plt.ylabel('Laying rate(%)')
plt.grid()
plt.show()
```
